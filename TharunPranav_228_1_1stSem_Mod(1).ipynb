    2. How many girls in the class
    3. How many Boys in the class
    4. Find ratio between Boys and Girls

```python
import pandas as pd

# Specify the file path
file_path = 'Data-analytic-course-assignment-sheet1.csv'

# Read the CSV file
df_Sdata = pd.read_csv(file_path)

# Strip whitespace and convert to lowercase for consistent formatting
df_Sdata['Gender'] = df_Sdata['Gender'].str.strip().str.lower()

# Recount the occurrences of each gender after cleaning
gender_counts_cleaned = df_Sdata['Gender'].value_counts()

# Extract counts for 'male' and 'female' after cleaning
total_boys_cleaned = gender_counts_cleaned.get('male', 0)
total_girls_cleaned = gender_counts_cleaned.get('female', 0)

print("Total no of girls in the class : ",total_girls_cleaned)
print("Total no of boys in the class : ",total_boys_cleaned)
print()

#------------------------------------------------------------------------------------------------------#

#4. Ratio between boys and girls

print("Ratio of Boys and Girls is given below :")

# Calculate the ratio
if total_boys_cleaned > 0:  # Ensure no division by zero
    ratio = total_girls_cleaned / total_boys_cleaned
    print(f"Ratio of Girls to Boys is: {total_girls_cleaned}:{total_boys_cleaned} or approximately {ratio:.2f}:1")
else:
    print("No boys in the dataset, cannot compute ratio.")
```

Consider the df_Sdata data frame and do the following

    5. Find Average height of Boys and Girls
    6. Average height of the whole class

```python
import pandas as pd

# Specify the file path
file_path = 'Data-analytic-course-assignment-sheet1.csv'

# Read the CSV file
df_Sdata = pd.read_csv(file_path)

# Ensure the Height column is numeric
df_Sdata['Height'] = pd.to_numeric(df_Sdata['Height'], errors='coerce')

# Filter boys and girls
boys_data = df_Sdata[df_Sdata['Gender'] == 'male']
girls_data = df_Sdata[df_Sdata['Gender'] == 'female']

# Calculate average height
avg_height_boys = boys_data['Height'].mean()
avg_height_girls = girls_data['Height'].mean()

print(f"Average height of boys: {avg_height_boys:.2f} cm")
print(f"Average height of girls: {avg_height_girls:.2f} cm")
print()

#------------------------------------------------------------------------------------------#

#6 To find the average height of the whole class

# Calculate the average height of the entire class
avg_height_class = df_Sdata['Height'].mean()

print("Average height of the class is given below :")
print(f"Average height of the whole class: {avg_height_class:.2f} cm")
```

    1. Define What is a Pandas Series

  A Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, etc.). It is similar to a column in a table or a 1D array. Each element in a Pandas Series has an associated index, allowing easy access to its elements.

    3. Key Difference Between Pandas DataFrame and Pandas Series
  >Pandas Series: A one-dimensional structure, similar to a single column or a 1D array.
    
  >Pandas DataFrame: A two-dimensional structure, analogous to a table, consisting of rows and columns, where each column is a Pandas Series.

    4. How df.iloc() and df.loc() used to access the data frame elements and give example

```python
#4 Difference between df.iloc() and df.loc()

import pandas as pd

# Creating a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}
df = pd.DataFrame(data, index=["R1", "R2", "R3"])

# Using .iloc[] to access data by position
print("Using iloc:")
print(df.iloc[1])  # Access the second row by index position
print(df.iloc[:, 1])  # Access the second column (Age)

# Using .loc[] to access data by labels
print("\nUsing loc:")
print(df.loc["R2"])  # Access the row with label 'R2'
print(df.loc[:, "City"])  # Access the 'City' column
```
